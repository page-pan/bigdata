```python
# 示例单元格
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 4))
plt.plot(x, y, 'b-', linewidth=2)
plt.title('RISE 幻灯片演示 - 中文标题测试')
plt.xlabel('X轴')
plt.ylabel('Y轴')
plt.grid(True)
plt.show()
```
